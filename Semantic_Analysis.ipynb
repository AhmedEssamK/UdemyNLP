```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
a = "This is a good movie"
```

```python
sid.polarity_scores(a)
```

```python
a = "This was the best, most awesome movie EVER MADE!!!"
```

```python
sid.polarity_scores(a)
```

```python
a = "This was the WORST movie that has ever disgraced the screen"
```

```python
sid.polarity_scores(a)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/amazonreviews.tsv', sep='\t')
```

```python
df.head()
```

```python
df['label'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
blanks = []
for i,lb,rv in df.itertuples():
  if type(rv) == str:
    if rv.isspace():
      blanks.append(i)
```

```python
df.drop(blanks, inplace=True)
```

```python
df.iloc[0]['review']
```

```python
sid.polarity_scores(df.iloc[0]['review'])
```

```python
df['scores'] = df['review'].apply(lambda review: sid.polarity_scores(review))
```

```python
df.head()
```

```python
df['compounds'] = df['scores'].apply(lambda d:d['compound'])
```

```python
df.head()
```

```python
df['comp_score'] = df['compounds'].apply(lambda score: 'pos' if score >=0 else 'neg')
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(df['label'], df['comp_score'])
```

```python
print(classification_report(df['label'], df['comp_score']))
```

```python
print(confusion_matrix(df['label'], df['comp_score']))
```
